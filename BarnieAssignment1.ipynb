```python
import math

def mod10(a, b):
    if a + b > 9:
        return (a + b) - 10
    else:
        return a + b

def chain(x, y, leng):
    fchain = [int(x), int(y)]
    for i in range(2, int(leng)):
        next = mod10(fchain[i-2], fchain[i-1])
        fchain = fchain + [int(next)]
    return fchain

def lpoint(fchain):
    repeat = []
    for j in range(2, len(fchain)):
        if (fchain[j] == fchain[0]) and (fchain[j+1] == fchain[1]):
            repeat = repeat + [j]
    return repeat

def add(fchain):
    if math.gcd(fchain[0], fchain[1]) == 1:
        return 60
    else:
        return fchain[0] + fchain[1]

ncoprime = 0
for j in range(1, 9):
    for i in range(1, 9):
        if math.gcd(j,i) == 1:
            ncoprime += 1

print(ncoprime)
test = chain(1, 5, 10)
test2 = chain(1,5,120)
test3 = chain(6,7,120)
test4 = chain(4,2,120)
test5 = chain(2,3,200)
test6 = chain(3,7,200)
test7 = chain(3,9,120)
test8 = chain(2,4,250)
test9 = chain(9,3,120)
test10 = chain(6,2,250)
test11 = chain(5,1,120)
test12 = chain(3,6,250)
lp2 = lpoint(test2)
lp3 = lpoint(test3)
lp4 = lpoint(test4)
lp5 = lpoint(test5)
lp6 = lpoint(test6)
lp7 = lpoint(test7)
lp8 = lpoint(test8)
lp9 = lpoint(test9)
lp10 = lpoint(test10)
lp11 = lpoint(test11)
lp12 = lpoint(test12)
add4 = add(test4)
add7 = add(test7)

print(test)
print(test2)
print(test11)
print(lp2)
print(lp3)
print(lp4)
print(lp5)
print(lp6)
print(lp7)
print(lp8)
print(lp9)
print(lp10)
print(lp11)
print(lp12)
print(test7)
print(test10)
print(test4)
```
